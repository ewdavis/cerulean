```python
import re, pprint

def getWordFreq(text):
    infile = open(text, "r", encoding='utf-8', errors='ignore')
    corpus = infile.readlines()
    wordList = []
    for line in corpus:
        line = re.sub("e'\W", "ed", line)
        line = re.sub("\W'", " ", line)
        line = re.sub("'\W", " ", line)
        line = re.sub("[^a-zA-Z']+", " ", line)
        line = line.lower()
        line = line.split()
        wordList = wordList + line
    wordFrequency = dict()
    for word in wordList:
        if (word == "ll"):
            word = "will"
        if (len(word) <= 1):
            continue
        if (word not in wordFrequency):
            wordFrequency[word] = 1
        else:
            wordFrequency[word] = wordFrequency[word] + 1
    sortedFrequency = sorted(wordFrequency.items(), key=lambda x: x[1], reverse=True)
    return(sortedFrequency)
```

```python
wordFreq = dict(getWordFreq("../data/faulkner/faulkner_absalom_absalom.txt"))
```

```python
for word in wordFreq.keys():
    if (re.search("^y.*", word)):
        print(word, "\t", wordFreq[word])
```
